#### Lambda Functions

Syntax: lambda arguments: expression

```python
def addition(a,b):
    return a + b

addition(2,3)
```

```python
add_two_values = lambda a,b: a+b
add_two_values(2,3)
```

```python
def even(num):
    if num%2 == 0:
        return True

even(30)
```

```python
even_no = lambda num: num % 2 == 0
even_no(20)
```

```python
# multiple parameters
add_multiple_values = lambda x,y,z: x+y+z
add_multiple_values(20,30,40)
```

#### Lambda + Map Function

```python
numbers = [1, 2, 3, 4, 5, 6]
def square(number):
    return number**2
square(2)
```

```python
square_numbers = list(map(lambda num: num**2, numbers))
square_numbers
```

#### Map Function in Python

```python
# Map to convert a list of strings to integers
# Use map to convert strings to integers

str_numbers = ['1', '2', '3', '4', '5']
int_numbers = list(map(int, str_numbers))

print(int_numbers)
```
